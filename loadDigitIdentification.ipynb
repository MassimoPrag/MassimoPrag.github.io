```python
#!pip install pandas scikit-learn matplotlib networkx
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import numpy as np

digits = datasets.load_digits() #load_digits(), sklearn preloaded data set, consists of 8x8 pixel grayscale images of handwritten digits, from 0 to 9.

kmeans = KMeans(n_clusters = 10, random_state = 42)
clusters = kmeans.fit_predict(digits.data)

print(clusters[403])

fig, ax = plt.subplots(2, 5, figsize = (8,8))
centers = kmeans.cluster_centers_.reshape(10, 8, 8)
for axi, center in zip(ax.flat,centers):
    axi.set(xticks=[],yticks = [])
    axi.imshow(center,interpolation = 'nearest', cmap=plt.cm.binary)


#elbow, run kmeans multiple times and plot the inertias, 
# not really useful in situation because only 10 different types of numbers 0-9, so 10 clusters
inertias = [] # inertias are the sum of the square distances to the nearest cluster center
ks = range(1,21)
for k in ks:
    kmeans = KMeans(n_clusters = k, random_state = 42, n_init = 10).fit(digits.data)
    inertias.append(kmeans.inertia_)

plt.figure(figsize = (8,8))
plt.plot(ks, inertias, marker = 'o')
plt.xlabel("K values")
plt.ylabel("Inertia values")
plt.show()


plt.figure(8, figsize=(3, 3))
plt.imshow(digits.images[0], cmap=plt.cm.gray_r, interpolation="nearest")#print out a selected number

```
